```python
import requests
from pprint import pprint
import json


# json.load, json.loads, json.dump, json.dumps


def request(url, headers, payload, request_type="GET"):
    response = requests.request(request_type, url, headers=headers, data = payload)
    return json.loads(response.text)
    
    
```

```python
summary_url = "https://api.covid19api.com/summary"
response = request(summary_url, headers={}, payload={})
```

```python
for country in response['Countries']:
    print(country['Country'])
```
